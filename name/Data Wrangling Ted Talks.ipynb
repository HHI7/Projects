```python
def get_top(feature):
    popular_talks = ted[['title', 'main_speaker', 'views','comments','film_date', 'published_date','duration']]\
                    .sort_values(by=feature,ascending=False).set_index(feature).reset_index().head(10)
    return popular_talks
```

```python
def get_graph(feature):
    ted_top = ted[['title', 'main_speaker', 'views','comments', 'published_date','duration']].sort_values(feature, ascending=False).head(10)
    x_labels=ted_top['title'].to_list()
    plt.figure(figsize=(12,8))
    ax=sns.barplot(x='title',y=feature,data=ted_top)
    ax.set_title('TOP 10 TED Talks by '+str(feature).upper())
    ax.set_xlabel('TITLE')
    ax.set_ylabel(feature)
    ax.set_xticklabels(x_labels,rotation='vertical')
    
    rects=ax.patches

    
    labels=ted_top[feature].to_list()
    
    for rect,label in zip(rects,labels):
        height=rect.get_height()
        ax.text(rect.get_x()+rect.get_width()/2,height,label,ha='center',va='bottom')
```
